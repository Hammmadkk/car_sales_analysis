```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/car_sales_data.csv', parse_dates=['Date'])
df.head()
```

```python
# Line Plot Example
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['Units Sold'], marker='o', linestyle='-', color='teal')
plt.title('Units Sold Over Time')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.grid(True)
plt.show()

obs_text = "###  Observation - Line Plot  Units Sold Over Time \n Units sold vary day to day, with values mostly between 1 and 5. \n There are frequent spikes at 5 units, suggesting peak performance days (likely marketing or dealership-driven). \n No obvious upward or downward trend  sales volume is inconsistent, indicating demand variability."
from IPython.display import Markdown
Markdown(obs_text)
```

```python
# 2. Scatter Plot - Units Sold vs Total Sales
plt.figure(figsize=(8, 5))
plt.scatter(df['Units Sold'], df['Total Sales'], color='green', alpha=0.6)
plt.title('Units Sold vs Total Sales')
plt.xlabel('Units Sold')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()

obs_text = "###  Observation - Scatter Plot  Units Sold vs Total Sales \n More units sold generally lead to higher total sales, but not always.\n A single unit sold can result in very high revenue  due to luxury models like Convertibles or Trucks. \n Indicates a mix of low-volume high-price and high-volume mid-price sales strategies."
from IPython.display import Markdown
Markdown(obs_text)
```

```python
# 3. Bar Chart - Total Sales by Region
region_sales = df.groupby('Region')['Total Sales'].sum().sort_values()
plt.figure(figsize=(8, 5))
region_sales.plot(kind='bar', color='coral')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales Amount')
plt.tight_layout()
plt.show()

obs_text = "###  Observation - Bar Chart  Total Sales by Region \n West region has the highest total sales  strong dealership network or customer base. \n East is also performing well, but slightly below the West. \n North and South have lower sales, suggesting an opportunity to improve regional performance or targeted marketing"
from IPython.display import Markdown
Markdown(obs_text)
```

```python
# 4. Histogram - Distribution of Unit Prices
plt.figure(figsize=(8, 5))
plt.hist(df['Unit Price'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Unit Prices')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

obs_text = "###  Observation - Histogram  Distribution of Unit Prices n- Most unit prices fall between 20,000 and 80,000, forming the bulk of the market.\n Some outliers above 80,000 indicate the presence of premium models. \n The company seems to focus more on mid-range vehicles but still caters to a high-end segment."
from IPython.display import Markdown
Markdown(obs_text)
```

```python
# 5. Pie Chart - Car Model Distribution
car_counts = df['Car Model'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(car_counts, labels=car_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Car Model Share')
plt.axis('equal')
plt.tight_layout()
plt.show()

obs_text = "###  Observation - Pie Chart  Car Model Share\n- SUVs and Sedans dominate the share  these are the most preferred models. \n Minivans and Trucks have moderate demand. \n Hatchbacks and Convertibles have smaller slices  niche or selective buyers. \n Suggests the business relies on core high-demand models with some diversity."
from IPython.display import Markdown
Markdown(obs_text)
```

```python
# 6. Subplot - Units Sold and Total Sales Over Time
fig, axs = plt.subplots(2, 1, figsize=(12, 8))
axs[0].plot(df['Date'], df['Units Sold'], color='blue', marker='o')
axs[0].set_title('Units Sold Over Time')
axs[1].plot(df['Date'], df['Total Sales'], color='darkgreen', marker='x')
axs[1].set_title('Total Sales Over Time')
plt.tight_layout()
plt.show()

obs_text = "###  Observation - Subplots  Units Sold & Total Sales Over Time \nNot always in sync: sometimes fewer units generate higher sales, showing pricing power. \n Certain spikes in total sales occur on days with only 12 units sold  indicating luxury car transactions."
from IPython.display import Markdown
Markdown(obs_text)
```

```python
# 7. Customized Dual Axis Plot - Units Sold and Total Sales
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.set_xlabel('Date')
ax1.set_ylabel('Units Sold', color='tab:red')
ax1.plot(df['Date'], df['Units Sold'], color='tab:red', label='Units Sold')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel('Total Sales', color='tab:blue')
ax2.plot(df['Date'], df['Total Sales'], color='tab:blue', label='Total Sales')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Units Sold and Total Sales Over Time')
fig.tight_layout()
plt.show()

obs_text = "###  Observation - Dual Axis Plot  Units Sold vs Total Sales \n- Selling just **1-2 units** can lead to high revenue (likely luxury cars). \n  Higher sales do **not always mean** higher revenue.\n Suggests importance of **model pricing** over quantity."
from IPython.display import Markdown
Markdown(obs_text)
```
